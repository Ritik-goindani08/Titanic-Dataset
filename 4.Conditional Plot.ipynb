```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Datasets/train.csv")
dataset.head()
```

```python
col = ["Survived","Pclass","Sex",'Age', 'SibSp','Parch','Fare','Embarked']
dataset = dataset[col]
dataset.head()
```

```python
dataset["Age"].value_counts().head()
```

```python
dataset["Age"].value_counts(bins = 10)
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(axis = 0 , inplace = True)
```

```python
dataset.info()
```

```python
sns.distplot(dataset["Age"])
```

```python
dataset.Age.value_counts(bins = 10, normalize = True)*100
```

```python
sns.kdeplot(dataset["Age"])
```

```python
sns.kdeplot(dataset["Age"], shade = True, color = "Red")
```

```python
g = sns.FacetGrid(dataset, col = "Survived", size = 6)
g.map(sns.kdeplot,"Age", shade = True, color = "green")
```

```python
g = sns.FacetGrid(dataset, col = "Survived", row = "Pclass", size=6)
g.map(sns.kdeplot,"Age", shade = True , color = "Red")
```

```python
g = sns.FacetGrid(dataset, col = "Survived", row = "Pclass", hue = "Sex" ,size = 5)
g.map(sns.kdeplot,"Age", shade = True)
g.add_legend()
sns.despine(left = True , bottom = True)
```

# style
1. whitegrid
2. darkgrid
3. ticks
4. white
5. dark

```python
g = sns.FacetGrid(dataset, col = "Survived" ,row = "Pclass",hue = "Sex", size = 4)
g.map(sns.kdeplot,"Age", shade = True)
g.add_legend()
sns.despine(left = True, bottom = True)
sns.set_style("whitegrid")
plt.show()
```

```python
g = sns.FacetGrid(dataset, col = "Survived" ,row = "Pclass",hue = "Sex", size = 4)
g.map(sns.kdeplot,"Age", shade = True)
g.add_legend()
sns.despine(left = True, bottom = True)
sns.set_style("darkgrid")
plt.show()
```

```python
g = sns.FacetGrid(dataset, col = "Survived" ,row = "Pclass",hue = "Sex", size = 4)
g.map(sns.kdeplot,"Age", shade = True)
g.add_legend()
sns.despine(left = True, bottom = True)
sns.set_style("ticks")
plt.show()
```

```python
g = sns.FacetGrid(dataset, col = "Survived" ,row = "Pclass",hue = "Sex", size = 4)
g.map(sns.kdeplot,"Age", shade = True)
g.add_legend()
sns.despine(left = True, bottom = True)
sns.set_style("white")
plt.show()
```

```python
g = sns.FacetGrid(dataset, col = "Survived" ,row = "Pclass",hue = "Sex", size = 4)
g.map(sns.kdeplot,"Age", shade = True)
g.add_legend()
sns.despine(left = True, bottom = True)
sns.set_style("dark")
plt.show()
```
